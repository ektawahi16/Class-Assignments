```python
import random

# A dictionary defined to decide who wins. Contestants are stored as key and winner and army number as value
# This dictionary is used by battle function to check which army's unit wins for every round.
who_wins = {('soldier', 'archer'): ('archer', 2),
            ('archer', 'soldier'): ('archer', 1),
            ('archer', 'archer'): 'tie',
            ('soldier', 'knight'): ('soldier', 1),
            ('knight', 'soldier'): ('soldier', 2),
            ('soldier', 'soldier'): 'tie',
            ('knight', 'archer'): ('knight', 1),
            ('archer', 'knight'): ('knight', 2),
            ('knight', 'knight'): 'tie',
            ('wizard', 'archer'): ('archer', 2),
            ('archer', 'wizard'): ('archer', 1),
            ('wizard', 'soldier'): ('wizard', 1),
            ('soldier', 'wizard'): ('wizard', 2),
            ('wizard', 'knight'): ('wizard', 1),
            ('knight', 'wizard'): ('wizard', 2),
            ('wizard', 'wizard'): 'tie'
            }

# Initializing a blank list for both armies
army_p = []
army_c = []

# Initializing wallet balance for both armies with $10
```

```python
"""
Function name       :   army_selection

Purpose             :   Show the user a menu of units available and facilitate selection. 
                        Stores the selected units in list variable passed as parameter and 
                        reduces money from wallet (wallet variable passed as parameter) with every purchase.
"""

def army_selection():
    wallet_p = 10
    wallet_c = 10
    exit_p = 0
    exit_c = 0
    
    # Loop until the exit flag goes to 1
    while exit_p == 0 or exit_c == 0:
        
        print("Make your selection -")
        print("1 - soldier    1\n2 - archer     2\n3 - knight     1\n4 - wizard     2\n5 - Ready for battle (exit)\n")
        
        if exit_p == 0:
            # Take input from user based on menu options
            selection = (input("Select your unit, PLAYER : "))

            # Check the input for every valid input and take suitable action
            if selection == '1':
                # based on the selection, corresponding unit is added to the end of the list
                army_p.append('soldier')
                # after adding unit, wallet balance is deduction as per the unit price.
                wallet_p -= 1
                print("Remaining wallet balance = ", wallet_p)

            elif selection == '2':
                army_p.append('archer')
                wallet_p -= 2
                print("Remaining wallet balance = ", wallet_p)

            elif selection == '3':
                army_p.append('knight')
                wallet_p -= 1
                print("Remaining wallet balance = ", wallet_p)

            elif selection == '4':
                army_p.append('wizard')
                wallet_p -= 2
                print("Remaining wallet balance = ", wallet_p)

            elif selection == '5':
                # Once the user chooses to exit the selection, final army is displayed.
                #print("PLAYER'S final army: ", army_name, "\n")
                # Exit flag is set to 1 to break the loop.
                exit_p = 1

            else:
                print("invalid selection. try again.")

        if exit_c == 0:
            print("Now, COMPUTER's turn to choose!")
            ar = ''
            cst = 0
            co = 5
            x = 1
            while(True):
                if(x == 1):
                    co=random.randint(1,4)
                else:
                    co=random.randint(1,5)
                
                x = x+1
                if co == 1:
                    ar = 'soldier'
                    cst = 1
                elif co == 2:
                    ar = 'archer'
                    cst = 2
                elif co == 3:
                    ar = 'knight'
                    cst = 1
                elif co == 4:
                    ar = 'wizard'
                    cst = 2
                elif co == 5:
                    exit_c = 1
                    break;

                if wallet_p>(wallet_c - cst):
                    print("Computer chose ", ar)
                    if wallet_c-cst>0:
                        army_c.append(ar)
                        wallet_c-=cst
                        break
                #else continue
            
        # At the end of each loop, check for wallet balance and set exit flag to 1 if wallet balance is 0.
        if wallet_p == 0:
            print("Your wallet is empty now!")
            #print("Your final army: ", army_p)
            exit_p = 1

        if wallet_c == 0:
            print("Computer's wallet empty!")
            exit_c = 1


"""
Function name       :   battle

expected parameters :   1. unit from army 1 for current round of battle
                        2. unit from army 2 for current round of battle

Purpose             :   This function accepts units from both armies as parameter. 
                        checks in the dictionary for a Key as combination of both units
                        which returns the winner as value. Part of this value is used to check
                        which army's unit wins and the unit from opponent army is removed from the its list.
                        if its a tie , both units are removed from corresponding lists. 
                        When units are removed, always 0th index is removed as both army's list variable's
                        0th index are set for battle in every round.
"""


def battle(unit_i, unit_ii):
    # print which units from each army are battling in current round
    print("\nArmy PLAYER:", unit_i, " vs Army COMPUTER:", unit_ii)
    # who_wins dictionary has a combination of all contestants and winner. winner returned as value is stored in winner.
    winner = who_wins[unit_i, unit_ii]

    # This checks which army's unit won. Opponent army's unit is removed from its list.
    if winner[1] == 1:
        # Displaying which army and which unit won the round.
        print("}}}Army PLAYER:", unit_i, "wins this round")
        # Removing the unit from opponent army.
        army_c.pop(0)

    if winner[1] == 2:
        print(">>>Army COMPUTER:", unit_ii, "wins this round")
        army_p.pop(0)

    # If its a tie, both units lose. Now both units are removed from their corresponding lists.
    if winner == 'tie':
        print("~~~its a tie, both units die.")
        army_p.pop(0)
        army_c.pop(0)
```

```python
army_selection()

# Both final armies are displayed
print("PLAYER army:", army_p)
print("COMPUTER army:", army_c)

print("\n######## Let the battle begin #######\n")

# The life of each army = no. of units alive = length of list variable (army_1 or army_2).
life_p = len(army_p)
life_c = len(army_c)

# Checking if either or both armies are empty and displaying appropriate message.
if life_p == 0 and life_c == 0:
    print("Both Armies are empty! No battle.")
    exit(1)
elif life_p == 0 and life_c > 0:
    print("Army Player is empty. Army Computer wins.")
    exit(1)
elif life_p > 0 and life_c == 0:
    print("Army Computer is empty. Army Player wins")
    exit(1)

    
# This loop simulates the battle by calling battle function. It loops until either or both armies are empty.
while len(army_p) > 0 and len(army_c) > 0:
    # 0th index unit in the army list of both armies battle in each round which is updated as per battle logic.
    battle(army_p[0], army_c[0])

if len(army_p) > 0:
    print("\n#### ---- ARMY PLAYER WINS! ---- #####")
if len(army_c) > 0:
    print("\n#### ---- ARMY COMPUTER WINS! ---- #####")
if len(army_p) == 0 and len(army_c) == 0:
    print("\n#### ---- BOTH ARMIES ARE DEAD, ITS A TIE ---- ####")
```
